###Classifiy the stations into 64 areas according to (latitude, longitude, average_days)

The following is the function to find the median of stations' latitude/longitude weighted by number of years that recorded both Tmax and Tmin, and then assign certain area code to the parted stations.

```python
def find_nearest(array,value):
    idx = numpy.argmin(numpy.abs(array-value))
    #idx = (numpy.abs(array-value)).argmin()
    return idx

def assign_area(total_data, key, area):
    index = total_data[:,3] == area
    data = total_data[index,:]
    #temp = numpy.resize(data[:,key], (1,data[:,key].shape[0]))
    #data_sorted = data[numpy.lexsort(temp)]
    data_sorted = data[numpy.argsort(data[:,key])]
    #print data_sorted
    data_cumm = numpy.asarray(data_sorted)
    data_cumm[:,0] = numpy.cumsum(data_sorted[:,0])
    #print data_cumm
    median =  data_cumm[data_cumm.shape[0]-1,0]/2
    idx = find_nearest(data_cumm[:,0], median)
    median_key = data_cumm[idx,key]
    #print data.shape, data_cumm[idx,:], median_key
    smaller = data[:,key] <= median_key
    larger = data[:,key] > median_key
    data[smaller, 3] = area*2
    data[larger, 3] = area*2+1
    total_data[index,:]=data
    return total_data
```

```python
tree = Djoined.loc[:,['years','latitude','longitude']].copy()
temp_data = tree.values
areas = numpy.zeros((temp_data.shape[0],1))
data = numpy.concatenate((temp_data, areas),axis=1)
index = numpy.isnan(data[:,0])
data[index,0] = 0
selected_data = data[(data[:,0]>0),:]
idx = (data[:,0]>0)
num_iterations = 8
areas = [0]
i = 0
for i in xrange(num_iterations):
    key = i % 2 +1
    for j in xrange(len(areas)-1,-1,-1):
        selected_data = assign_area(selected_data,key, areas[j])
    new = areas[len(areas)-1]
    for j in xrange(len(areas)):
        new = new+1
        areas.append(new)
    #print areas
for i in xrange(len(areas)):
    index = selected_data[:,3] == i
    temp_data = selected_data[index,:]
    min_latitude = numpy.min(temp_data[:,1])
    max_latitude = numpy.max(temp_data[:,1])
    min_longitude = numpy.min(temp_data[:,2])
    max_longitude = numpy.max(temp_data[:,2])
    s = temp_data.shape[0]
    print 'area %d: latitude range: (%f, %f), longitude range(%f, %f), number: %d'%(i,min_latitude, max_latitude, min_longitude, max_longitude, s)
data[idx,:] = selected_data
data[~idx,3] = len(areas)
s = data[~idx,3].shape[0]
print 'area %d: %d stations with no eligible measurements' %(len(areas),s)
```

From above, we can see that the eligible stations are partitioned into 256 areas, and in each area, the number of stations are similar, but not the same. That is because I use the weighted median instead of median. Number of years is a factor affecting the partition. The stations that are not selected are assinged to area 256 so that it is easier to do PCA in the next step.

```python
Djoined['area'] = data[:,3]
cols = Djoined.columns.tolist()
Djoined = Djoined[cols]
Djoined.head()
```

I also plot the stations on the earth map, to verify the partition result. 

```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
from scipy.interpolate import griddata

lonmin=-180;lonmax=180;latsmin=-80;latsmax=80;
plt.figure(figsize=(15,10),dpi=300)
m = Basemap(projection='merc',llcrnrlat=latsmin,urcrnrlat=latsmax,\
            llcrnrlon=lonmin,urcrnrlon=lonmax,lat_ts=20,resolution='i')
m.drawcoastlines(linewidth=0.5)#,color='grey')

parallels = np.arange(-80,81,10.)
m.drawparallels(parallels,labels=[False,True,True,False])
meridians = np.arange(10.,351.,20.)
m.drawmeridians(meridians,labels=[True,False,False,True])

lats= Djoined.loc[:,'latitude'].values
lons= Djoined.loc[:,'longitude'].values
ll=len(lons)
selection = Djoined.loc[:,'area'].values<256
lats = lats[selection]
lons = lons[selection]
x, y = m(lons,lats)
z= Djoined.loc[:,'area'].values
z = z[selection]
m.scatter(x, y, c=z, s=2, cmap=plt.cm.jet, edgecolors='None')
cbar = plt.colorbar()
#plt.clim(-16,32)
cbar.ax.set_ylabel('area index')
plt.title('Eligible weather stations')
plt.show()
```

Only area 0-255 is plotted, that is, the stations that are not eligible are not shown in the picture. The plot shown is very intuitive, for example, in America, the stations are very dense and the each station records many years containing Tmax and Tmin, so there are many areas in America.
