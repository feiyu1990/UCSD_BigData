```python
%%writefile find_tmax_with_years.py
#!/usr/bin/python
"""
count the number of measurements of each type
"""
import sys
sys.path.append('/usr/lib/python2.6/dist-packages')
from mrjob.job import MRJob
import re
import pandas as pd
import numpy as np
from sys import stderr

WORD_RE = re.compile(r"[\w']+")

#logfile=open('log','w')
#logfile=stderr

class MRWeather(MRJob):

    def my_mapper(self, _, line):
        self.increment_counter('MrJob Counters','my_mapper',1)
        elements=line.split(',')
        #logfile.write('%s\n' % type(line))
        if elements[1]=='TMAX':
            key = (elements[0],elements[2])
            n_days = sum([e!='' for e in elements[3:]])                
            value = []
            for each in elements[3:]:
                if each == '':
                    value.append(np.nan)
                else:
                    value.append(float(each))
            s = pd.Series(value)
            s = s.interpolate()
            #s = s.fillna(s.mean())
            n_days = sum([e!='' for e in elements[3:]])
            all_days = sum([not np.isnan(e) for e in s])
            if n_days>200 and all_days ==365:
                t_average = float(sum([e for e in s]))/365
                yield key[0], (key[1], t_average)
                
    def my_reducer(self,word,counts):
        self.increment_counter('MrJob Counters','my_reducer',1)
        sum2 = 0;sum1=0
        ten_years_previous = [];averages_previous = []
        ten_years_recent = [];averages_recent = []
        for year, t_average in counts:
            ten_year = int(year)
            if ten_year >=2000:
                averages_recent.append(t_average)
                ten_years_recent.append(ten_year)
                sum1 = sum1 + 1
            if ten_year <=1980 or ten_year>=1951:
                averages_previous.append(t_average)
                ten_years_previous.append(ten_year)                
                sum2 = sum2 + 1
        if len(ten_years_previous) and len(ten_years_recent) and (sum1>10) and (sum2>10):
            average_previous = sum(averages_previous)/len(averages_previous)
            average_recent = sum(averages_recent)/len(averages_recent)
            yield word,(average_previous,average_recent, sum1+sum2) 

    def steps(self):
        return [self.mr(mapper=self.my_mapper,reducer = self.my_reducer)]

if __name__ == '__main__':
    MRWeather.run()
```

```python
def find_nearest(array,value):
    idx = numpy.argmin(numpy.abs(array-value))
    #idx = (numpy.abs(array-value)).argmin()
    return idx

def assign_area(total_data, key, area):
    index = total_data[:,3] == area
    data = total_data[index,:]
    #temp = numpy.resize(data[:,key], (1,data[:,key].shape[0]))
    #data_sorted = data[numpy.lexsort(temp)]
    data_sorted = data[numpy.argsort(data[:,key])]
    #print data_sorted
    data_cumm = numpy.asarray(data_sorted)
    data_cumm[:,0] = numpy.cumsum(data_sorted[:,0])
    #print data_cumm
    median =  data_cumm[data_cumm.shape[0]-1,0]/2
    idx = find_nearest(data_cumm[:,0], median)
    median_key = data_cumm[idx,key]
    #print data.shape, data_cumm[idx,:], median_key
    smaller = data[:,key] <= median_key
    larger = data[:,key] > median_key
    data[smaller, 3] = area*2
    data[larger, 3] = area*2+1
    total_data[index,:]=data
    return total_data
```

```python
tree = Djoined.loc[:,['years','latitude','longitude']].copy()
temp_data = tree.values
areas = numpy.zeros((temp_data.shape[0],1))
data = numpy.concatenate((temp_data, areas),axis=1)
index = numpy.isnan(data[:,0])
data[index,0] = 0
selected_data = data[(data[:,0]>0),:]
idx = (data[:,0]>0)
num_iterations = 9
areas = [0]
i = 0
for i in xrange(num_iterations):
    key = i % 2 +1
    for j in xrange(len(areas)-1,-1,-1):
        selected_data = assign_area(selected_data,key, areas[j])
    new = areas[len(areas)-1]
    for j in xrange(len(areas)):
        new = new+1
        areas.append(new)
    #print areas
for i in xrange(len(areas)):
    index = selected_data[:,3] == i
    temp_data = selected_data[index,:]
    min_latitude = numpy.min(temp_data[:,1])
    max_latitude = numpy.max(temp_data[:,1])
    min_longitude = numpy.min(temp_data[:,2])
    max_longitude = numpy.max(temp_data[:,2])
    s = temp_data.shape[0]
    print 'area %d: latitude range: (%f, %f), longitude range(%f, %f), number: %d'%(i,min_latitude, max_latitude, min_longitude, max_longitude, s)
data[idx,:] = selected_data
data[~idx,3] = len(areas)
s = data[~idx,3].shape[0]
print 'area %d: %d stations with no eligible measurements' %(len(areas),s)
```

```python
%%writefile find_tmax.py
#!/usr/bin/python
"""
count the number of measurements of each type
"""
import sys
sys.path.append('/usr/lib/python2.6/dist-packages')
from mrjob.job import MRJob
import re
import pandas as pd
import numpy as np
from sys import stderr

WORD_RE = re.compile(r"[\w']+")

#logfile=open('log','w')
#logfile=stderr

class MRWeather(MRJob):

    def my_mapper(self, _, line):
        self.increment_counter('MrJob Counters','my_mapper',1)
        elements=line.split(',')
        #logfile.write('%s\n' % type(line))
        if elements[1]=='TMAX':
            key = (elements[0],elements[2])
            n_days = sum([e!='' for e in elements[3:]])                
            value = []
            for each in elements[3:]:
                if each == '':
                    value.append(np.nan)
                else:
                    value.append(float(each))
            s = pd.Series(value)
            s = s.interpolate()
            #s = s.fillna(s.mean())
            n_days = sum([e!='' for e in elements[3:]])
            all_days = sum([not np.isnan(e) for e in s])
            if n_days>200 and all_days ==365:
                t_average = float(sum([e for e in s]))/365
                yield key[0], (key[1], t_average)
                
    def my_reducer(self,word,counts):
        self.increment_counter('MrJob Counters','my_reducer',1)
        sum2 = 0
        ten_years_previous = [];averages_previous = []
        ten_years_recent = [];averages_recent = []
        for year, t_average in counts:
            ten_year = int((int(year)-1763)/25)
            if ten_year ==9:
                averages_recent.append(t_average)
                ten_years_recent.append(ten_year)
                sum2 = sum2 + 1
            if ten_year ==7 or ten_year==6:
                averages_previous.append(t_average)
                ten_years_previous.append(ten_year)                
                sum2 = sum2 + 1
        if len(ten_years_previous) and len(ten_years_recent) and (sum2>15):
            average_previous = sum(averages_previous)/len(averages_previous)
            average_recent = sum(averages_recent)/len(averages_recent)
            yield word,(average_previous,average_recent) 

    def steps(self):
        return [self.mr(mapper=self.my_mapper,reducer = self.my_reducer)]

if __name__ == '__main__':
    MRWeather.run()
```
